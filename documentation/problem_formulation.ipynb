# Urban Commuter Mode Choice in Cycling vs. Transit Cities
## A Comparative Discrete Choice Analysis of Amsterdam and London


---

> **Abstract.** This study examines how commuters in London and Amsterdam choose between bus, metro, car, and bicycle. Using revealed preference data from the UK National Travel Survey (NTS) and the Dutch national travel survey ODiN 2023.

---

## 1. Introduction and Problem Formulation

Urban commuters face a daily decision: car, public transport, or bicycle. The outcome of this decision, aggregated across millions of individuals, determines congestion levels, carbon emissions, and the return on public infrastructure investment. Understanding *why* travellers choose a particular mode — and whether those drivers differ across cities with fundamentally different transport cultures — is therefore a first-order policy question (Ortúzar and Willumsen, 2011).

Amsterdam and London represent two archetypal European mobility regimes. Amsterdam is a cycling-dominant city where the bicycle accounts for roughly 60 % of commuting trips within the municipal boundary, supported by dense cycling infrastructure and flat topography (KiM, 2023). London, by contrast, is a transit-dominant metropolis whose commuters rely heavily on the Underground, bus network, and surface rail, with cycling remaining a minority mode despite recent investment (Department for Transport, 2024). Comparing these two cities allows us to test whether the same Level-of-Service (LoS) and socioeconomic attributes produce different behavioural responses under different urban transport regimes.

The analysis draws on two independent national travel surveys collected for the same calendar year: the Dutch *Onderweg in Nederland* (ODiN) 2023 and the UK *National Travel Survey* (NTS) 2023. Both are household-based, diary-style instruments that record individual trip attributes — mode, travel time, distance, purpose — alongside rich socioeconomic covariates. By harmonising these two sources into a shared schema, we obtain a cross-national dataset suitable for discrete choice modelling.

## 2. Research Question and Hypotheses

The central research question is:

> *What are the key determinants of commuter mode choice (car, public transport, bicycle) in Amsterdam and London, and do these determinants differ significantly between the two cities?*

Two testable hypotheses follow from the literature on random utility models and urban transport behaviour.

**Hypothesis 1 — Travel Time Sensitivity.** Travel time is a statistically significant negative predictor of public-transport and bicycle choice in both cities, but its marginal disutility is larger in London than in Amsterdam. The rationale is that London's greater spatial extent produces longer and more variable commutes, amplifying the perceived cost of each additional minute (Wardman, Chintakayala and de Jong, 2016).

Operationally, this hypothesis is tested by comparing the estimated travel-time coefficient $\hat{\beta}_{\text{time}}$ across city-specific Multinomial Logit (MNL) models and inspecting whether their confidence intervals overlap.

**Hypothesis 2 — Car Ownership Effect.** Car ownership and household income positively predict car choice in both cities, but the effect is significantly weaker in Amsterdam, where cycling provides a high-quality substitute that is accessible regardless of income (Harms, Bertolini and Brömmelstroet, 2014). 

This is tested via the magnitude and significance of $\hat{\beta}_{\text{car}}$ and $\hat{\beta}_{\text{inc}}$ in each city model.

## 3. Theoretical Framework — Random Utility Maximisation

The modelling backbone is the Random Utility Maximisation (RUM) framework formalised by McFadden (1974). Each commuter $n$ faces a choice set $C = \{\text{car},\; \text{PT},\; \text{bicycle}\}$ and selects the mode $i$ that maximises perceived utility:

$$
U_{ni} = V_{ni} + \varepsilon_{ni}
$$

Here $V_{ni}$ is the systematic (observable) component and $\varepsilon_{ni}$ is a random error term capturing unobserved taste variation. Under the assumption that the error terms are independently and identically distributed according to a Type-I Extreme Value (Gumbel) distribution, the choice probability takes the closed-form Multinomial Logit (MNL) expression (Train, 2009):

$$
P_{ni} = \frac{\exp(V_{ni})}{\sum_{j \in C} \exp(V_{nj})}
$$

This expression gives the probability that individual $n$ chooses mode $i$, conditional on the systematic utilities of all alternatives in $C$.

The systematic utility for mode $i$ chosen by individual $n$ is specified as a linear-in-parameters function:

$$
V_{ni} = \text{ASC}_i + \beta_{\text{time}} \cdot \text{TravelTime}_{ni} + \beta_{\text{inc}} \cdot \text{Income}_{n} + \beta_{\text{car}} \cdot \text{CarOwn}_{n} + \beta_{\text{age}} \cdot \text{Age}_{n} + \beta_{\text{gender}} \cdot \text{Gender}_{n}
$$

The alternative-specific constants $\text{ASC}_i$ absorb the average effect of all unobserved attributes for mode $i$ relative to the reference mode (car). Travel time enters as a generic coefficient because its disutility is hypothesised to be mode-invariant in our base specification; income, car ownership, age, and gender enter as individual-specific socioeconomic attributes interacted with the alternative-specific constants. Distance is excluded from the specification to avoid collinearity with travel time, which is the standard practice when both are available from the same trip record (Ben-Akiva and Lerman, 1985).

The model parameters $\boldsymbol{\theta} = \{\text{ASC}_i, \beta_{\text{time}}, \beta_{\text{inc}}, \beta_{\text{car}}, \beta_{\text{age}}, \beta_{\text{gender}}\}$ are estimated by Maximum Likelihood Estimation (MLE). The log-likelihood function for a sample of $N$ observed choices is:

$$
\mathcal{LL}(\boldsymbol{\theta}) = \sum_{n=1}^{N} \sum_{i \in C} y_{ni} \ln P_{ni}
$$

where $y_{ni} = 1$ if individual $n$ chose mode $i$ and $0$ otherwise. Goodness-of-fit is assessed via McFadden's pseudo-$R^2$:

$$
\rho^2 = 1 - \frac{\mathcal{LL}(\hat{\boldsymbol{\theta}})}{\mathcal{LL}(\mathbf{0})}
$$

where $\mathcal{LL}(\mathbf{0})$ is the log-likelihood of the equal-shares (null) model. Values of $\rho^2$ between 0.2 and 0.4 indicate excellent fit in discrete choice contexts (McFadden, 1978).

A key limitation of MNL is the Independence of Irrelevant Alternatives (IIA) property, which assumes that the ratio of choice probabilities between any two modes is unaffected by the attributes of a third mode. For the three-mode urban commute choice set considered here, IIA is a reasonable working assumption, though a Nested Logit extension grouping motorised modes (car + PT) against bicycle could be explored as a robustness check (Train, 2009).
